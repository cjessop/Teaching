# Classes and Object-Oriented Programming (OOP)


When describing physical objects, we usually reference its `attributes` - colour, material, weight etc. Certain attributes only apply to certain objects -> Modelling an object in code we decide what data is required to represent an object and what operations they can perform. Referred to as an objects    `state` and `behaviour`.

A simple example below of a class

````python
class Switch():
    def __init__(self):
        self.state = False

    def turnOn(self):
        self.state = True
        print("Switched On")

    def turnOff(self):
        self.state = False
        print("Switched Off")
````

To create an object using the class you would write it as:

````python
objSwitch = Switch()
````

What this says is find the `Switch` class, create a `Switch` object using that class, and then assign the resultant object to the variable `objSwitch`

Within the body of each class, you can then define any number of functions you require. Each function is a `behaviour` that an object can perform. Functions within classes are called `methods` -> "A function defined inside a class, always has at least one argument, most commonly `self`"

## Instantiation

When creating a class, the first `method` should be the `__init__` definition. When you create an object that uses that class, this method will automatically be run. It therefore makes the most sense that any initialisation code. Note that the `__init__` method is a reserved by the kernel

## Scope

Aside from the `global` and `local` scopes typically encountered, classes introduce the `object` or `instance` scope. `Methods` can use both `local` and `instance` variables, and any variable that does not begin with `self.` is a `local` variable.

To use a method associated with an object, the syntax is:

`<object>.<methodName>(<arguments>)`

Using the already defined class:

````python
objSwitch.turnOn()
objSwitch.turnOff()
````

````python
class Calculator:
    def add(self, num1, num2):
        return num1 + num2

    def subtract(self, num1, num2):
        return num1 - num2

    def multiply(self, num1, num2):
        return num1 * num2

    def divide(self, num1, num2):
        if num2 != 0:
            return num1 / num2
        else:
            print("Error: Division by zero is not allowed.")

# Example usage:
calculator = Calculator()
result = calculator.add(5, 3)
print(result)  # Output: 8

result = calculator.subtract(10, 4)
print(result)  # Output: 6

result = calculator.multiply(2, 6)
print(result)  # Output: 12

result = calculator.divide(10, 2)
print(result)  # Output: 5

result = calculator.divide(10, 0)
# Output: Error: Division by zero is not allowed.
````

# Encapsulation


The practice of bundling an object's data (attributes) and its related behavior (methods) into a single unit, providing controlled access to the internal workings
Public attributes and methods are defined in the regular way, such as self.value. Protected attributes are accessible only from within the class and subclasses, but not from outside and are defined with a leading single underscore `_`. Private attributes are hidden from outside of the class and can only be modified by methods within the same class. These are defined with two leading underscores `__`.

````python
class BankAccount:
    def __init__(self, account_number, balance, protected_value):
        self.__account_number = account_number  # Initialize the account number as a private attribute
        self.__balance = balance  # Initialize the account balance as a private attribute
        self._protected_value = protected_value  # Initialize the protected attribute

    def deposit(self, amount):
        self.__balance += amount  # Add the deposited amount to the balance

    def withdraw(self, amount):
        if amount <= self.__balance:  # Check if the withdrawal amount is less than or equal to the balance
            self.__balance -= amount  # Subtract the withdrawal amount from the balance
        else:
            print("Insufficient funds")  # Print a message if there are insufficient funds

    def get_balance(self):
        return self.__balance  # Return the current balance

    def __calculate_interest(self):
        interest_rate = 0.05  # Set the interest rate
        interest = self.__balance * interest_rate  # Calculate the interest based on the balance
        return interest  # Return the calculated interest

    def add_interest(self):
        interest = self.__calculate_interest()  # Calculate the interest
        self.__balance += interest  # Add the interest to the balance

# Example usage:
account = BankAccount("123456789", 1000, "protected_value")  # Create a BankAccount object with an account number, initial balance, and protected value

print(account.get_balance())  # Output: 1000

account.deposit(500)  # Deposit 500 into the account
print(account.get_balance())  # Output: 1500

account.withdraw(200)  # Withdraw 200 from the account
print(account.get_balance())  # Output: 1300

account.add_interest()  # Add interest to the account
print(account.get_balance())  # Output: 1365

print(account._protected_value)  # Output: protected_value
````

The bank account class shows the "Getter and Setter" techniques `get_balance()` and `set_balance()`. These methods provide controlled access to private attributes. `get_balance()` retrieves the balance without exposing the raw attribute, while `set_balance(amount)` ensures valid changes.



# Static and Class Methods

While regular methods operate on class instances, `static` methods belong directly to the class itself. They don't need an instance to be called and can't access instance attributes. They are utility functions attached to the class.

`class` methods have access to the class itself (but not specific instances), often used for factory methods or alternative constructors

````python
class Circle:
    pi = 3.14159  # Class attribute

    def __init__(self, radius):
        self.radius = radius

    @staticmethod
    def area_from_diameter(diameter):
        radius = diameter / 2
        return Circle.pi * radius**2  # Access class attribute directly

    @classmethod
    def from_diameter(cls, diameter):
        radius = diameter / 2
        return cls(radius)  # Create instance using class method

circle1 = Circle(5)
area1 = circle1.area_from_diameter(10)  # Accessing static method through instance
area2 = Circle.area_from_diameter(10)  # Accessing static method directly

circle2 = Circle.from_diameter(12)  # Creating instance using class method

print("Area1:", area1)
print("Area2:", area2)
print("Circle2 radius:", circle2.radius)
````

- **Static methods** 
  - Bound to the class, not individual instances
  - Defined using the `@staticmethod` decorator
  - Don't take `self` or `cls` as the first argument.
  - Cannot access or modify instance attributes or class attributes
  - Commonly used for:
    - Utility functions related to the class but not requiring instance data (e.g., mathematical calculations, conversions)
    - Organizing code and avoiding unnecessary instance creation

- **Class methods** 
  - Bound to the class, but can access class attributes and create new instances
  - Take `cls` as the first argument, referring to the class itself
  - Defined using the `@classmethod`decorator
  - Can access and modify class attributes, but not instance attributes directly
  - Commonly used for:
    - Alternative constructors, providing different ways to create instances
    - Factory methods, generating specific subclasses based on conditions
    - Accessing and modifying class-level data or metadata


## Polymorhpism


`Polymorphism` refers to a set of objects of different types sharing common behaviour, even when they do it in slightly different ways.
- **Benefits** 
  - Flexible Design: Code becomes more adaptable and can easily incorporate new types of objects in the future
  - Cleaner: Avoid duplicating code for similar tasks across different object types
  - Reusable: Can write general-purpose functions that work with a variety of objects without knowing their exact type


````python
class Shape: # Parent class
    def draw(self):
        raise NotImplementedError("Subclasses must implement the draw() method") # Section 1 

class Circle(Shape): # Child class # Section 2
    def __init__(self, radius):
        self.radius = radius

    def draw(self):
        print("Drawing a circle with radius", self.radius)

class Rectangle(Shape): # Child class
    def __init__(self, width, height):
        self.width = width
        self.height = height

    def draw(self):
        print("Drawing a rectangle with width", self.width, "and height", self.height) 

def draw_all_shapes(shapes): # Function that takes a list of shapes # Section 3
    for shape in shapes:
        shape.draw()  # Polymorphic call: calls the appropriate draw() method based on the object's type

# Create some shapes
circle = Circle(5)
rectangle = Rectangle(4, 6)

# Put them in a list
shapes = [circle, rectangle]

# Draw all shapes using polymorphism
draw_all_shapes(shapes) # Output: Drawing a circle with radius 5
````

### Step by Step Explanation of above


- **1 - Shape Base Class** 
  - Defines a common interface with the `draw()` method 
  - Error handling
- **2 - Circle and Rectangle Child Classes**
  - Example of Inheritence from the Base class, overrides the `draw()` method with their own object specific logic
- **3 - `draw_all_shapes` Function**
  - Takes in a list of `shape` objects
  - Iterates through the list and calls the `draw()` method on each object within the list 
  - The `draw()` method successfully runs on each object type without the function explicity knowing the type of each object

    - add(operand): Adds operand to the current result and updates current_result
    - subtract(operand): Subtracts operand from the current result and updates current_result

    - divide(operand): Divides the current result by operand and updates current_result
    - set_operator(op): Sets the current operator to op (e.g., '+', '-', etc.)

````python
class Calculator:
    def __init__(self):
        self.current_result = 0
        self.operator = None

    def add(self, operand):
        self.current_result += operand

    def subtract(self, operand):
        self.current_result -= operand

    def multiply(self, operand):
        self.current_result *= operand

    def divide(self, operand):
        if operand == 0:
            raise ZeroDivisionError("Cannot divide by zero")
        self.current_result /= operand

    def set_operator(self, op):
        self.operator = op

    def clear(self):
        self.current_result = 0
        self.operator = None

    def get_result(self):
        return self.current_result
````
